```python
from pandas_datareader import data
yahoo = data.DataReader('GOOG', start='2004', end='2019', data_source='yahoo')
print(yahoo.head())
print(yahoo.shape)   # (3617, 6)
```

```python
p_close = yahoo['Close']
# print(p_close)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
plt.figure(figsize=(10,7))
p_close.plot()
```

```python
plt.figure(figsize=(10,8))
p_close.plot(alpha=0.5, style='-')

respl = p_close.resample('BA').mean()
asfre = p_close.asfreq('BA')
respl.plot(style=':', color='red')
asfre.plot(style='--',color='yellow')
plt.legend(['input', 'resample', 'asfreq'], loc='upper left')
```

```python
fig, ax = plt.subplots(2, sharex=True)
data = p_close.iloc[:20]
data.asfreq('D').plot(ax=ax[0], marker='o')
data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='ffill').plot(ax=ax[1], style='--o')
ax[1].legend(['back-fill', 'forward-fill'])
```

```python
fig, axe = plt.subplots(3, sharex=True)

# 对数据应用时间频率，向后填充解决缺失值
p_close = p_close.asfreq('D', method='pad')

p_close.plot(ax=axe[0])
p_close.shift(900).plot(ax=axe[1])
p_close.tshift(900).plot(ax=axe[2])

# 设置图例与标签
local_max = pd.to_datetime('2007-11-05')
offset = pd.Timedelta(900, 'D')
axe[0].legend(['input'], loc=2)
axe[0].get_xticklabels()[4].set(weight='heavy', color='red')
axe[0].axvline(local_max, alpha=0.3, color='red')

axe[1].legend(['shift900'], loc=2)
axe[1].get_xticklabels()[4].set(weight='heavy', color='red')
axe[1].axvline(local_max + offset, alpha=0.3, color='red')

axe[2].legend(['tshift900'], loc=2)
axe[2].get_xticklabels()[1].set(weight='heavy', color='red')
axe[2].axvline(local_max + offset, alpha=0.3, color='red')
```

```python
ROI = 100 * (p_close.tshift(-365) / p_close -1)
ROI.plot()
plt.ylabel('% return on investment')
```

```python
# 移动时间窗口rolling
plt.figure(figsize=(10,10))
rolling = p_close.rolling(365, center=True)
data = pd.DataFrame({'input': p_close,
                    'one-year rolling_mean': rolling.mean(),
                    'one-year rolling_std': rolling.std()})

ax = data.plot(style=['-', '--', ':'], color=['blue', 'yellow', 'red'])
ax.lines[0].set_alpha(0.4)
```
