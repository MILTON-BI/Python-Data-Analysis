# 主成分分析PCA

- PCA是应用最广的无监督算法之一
- 虽然PCA是一种非常基础的降维算法，但它仍然是一个非常有用的工具
    - 尤其适用于数据可视化、噪音过滤、特征抽取和特征工程等领域

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

## 1. 主成分分析PCA简介

- 主成分分析是一个快速灵活的数据降维无监督方法

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
```

- 与回归分析中希望根据 x 值预测 y 值的思路不同， 无监督学习希望探索 x 值和 y 值之间的相关性
- 在主成分分析中， 一种量化两变量间关系的方法是在数据中找到一组主轴， 并用这些主轴来描述数据集

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

- 该拟合从数据中学习到了一些指标， 其中最重要的是“成分”和“可解释差异”

```python
print(pca.components_)
print(pca.explained_variance_)
```

```python
# 通过图形来查看上述参数的含义
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops = dict(arrowstyle='->', linewidth=2, shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# 画出数据
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal')
```

- 这些向量表示数据主轴，箭头长度表示输入数据中各个轴的“重要程度”
    - 更准确地说， 它衡量了数据投影到主轴上的方差的大小
- 每个数据点在主轴上的投影就是数据的“主成分”
- 这种从数据的坐标轴变换到主轴的变换是一个仿射变换
    - 仿射变换包含平移（translation）、旋转（rotation）和均匀缩放（uniform scaling）三个步骤

## 2. 主成分分析PCA的应用示例
### 2.1 用PCA降维
    
- 用 PCA 降维意味着去除一个或多个最小主成分， 从而得到一个更低维度且保留最大数据方差的数据投影
- PCA 降维的含义：
    - 沿着最不重要的主轴的信息都被去除了，仅留下了含有最高方差值的数据成分。 
    - 被去除的那一小部分方差值（与主轴上分布的点成比例）基本可以看成是数据在降维后损失的“信息”量
    - 这种降维后的数据集在某种程度上足以体现数据中最主要的关系：虽然大部分数据维度被削减，但数据的总体关系仍然被大致保留了下来

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print('original shape', X.shape)
print('transformed shape', X_pca.shape)
```

```python
# 进行数据降维的逆变换， 并且与原始数据一起画出
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.3, color='b')
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8, color='r')
plt.axis('equal')
```

### 2.2 用PCA做数据可视化：手写数字

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
# 用PCA将数据从64维投影到2维
pca = PCA(2)
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
# 画出每个点的前面两个主成分，更好地了解数据
plt.scatter(projected[:, 0], projected[:, 1], c=digits.target, edgecolors='none', alpha=0.5,
           cmap='viridis')
plt.xlabel('component one')
plt.ylabel('component two')
```

- 这些成分的含义： 整个数据是一个 64 维的点云，而且这些点还是每个数据点沿着最大方差方向的投影
- 我们找到了在 64 维空间中最优的延伸和旋转方案， 使得我们可以看到这些点在二维平面的布局
- 上述工作都是以无监督的方式进行的， 即没有参考标签

### 2.3 成分的含义

- PCA 可以被认为是选择最优基函数的过程， 这样将这些基函数中前几个加起来就足以重构数据集中的大部分元素。 
- 用低维形式表现数据的主成分， 其实就是与序列每一个元素相乘的系数
- 与像素基不同， PCA 基可以通过为一个均值加上 8 个成分， 来恢复输入图像最显著的特征
- 这就是 PCA 提供数据的低维表示的原理： 它发现一组比原始的像素基向量更能有效表示输入数据的基函数

### 2.4 选择成分的数量

- 可以将累计方差贡献率看作是关于成分数量的函数， 从而确定所需成分的数量

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

## 3. 用pca做噪声过滤

- 任何成分的方差都远大于噪音的方差， 所以相比于噪音， 成分应该相对不受影响。 
    - 因此，如果用主成份的最大子集重构该数据，可以实现选择性保留信号并且丢弃噪音。

```python
# 首先选几个无噪音的输入数据绘图
def plot_digits(data):
    fig, ax = plt.subplots(4, 10, figsize=(10, 4), subplot_kw={'xticks': [], 'yticks': []},
                          gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, axi in enumerate(ax.flat):
        axi.imshow(data[i].reshape(8, 8), cmap='binary', interpolation='nearest', clim=(0, 16))
plot_digits(digits.data)
```

```python
# 添加一些噪音并创建噪音数据集
np.random.seed(42)
noisy = np.random.normal(loc=digits.data, scale=4)
plot_digits(noisy)
```

```python
# 用噪音数据训练一个PCA，要求投影后保存50%的方差
pca = PCA(0.5).fit(noisy)
pca.n_components_     # 输出12，即这里50%的方差对应 12 个主成份
```

```python
# 计算这些主成分，然后用逆变换重构规律后的手写数字
components = pca.transform(noisy)
print(components)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

## 4. 脸特征

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

- 因为这是一个非常大的数据集， 所以我们将利用 RandomizedPCA。 
- 它包含了一个随机方法来估计前 N 个主成分， 比标准的 PCA 评估器速度更快， 并且特别适用于高维数据（这里的维度将近 3000）。
- 先来看看前 150 个成分

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(150)
pca.fit(faces.data)
```

- 将这个例子中带有前面几个主成分的图像可视化。
- 这些成分被称作“特征向量”， 因此这些图像的类型通常被称作“特征脸”

```python
fig, ax = plt.subplots(3, 8, figsize=(9,4), 
                       subplot_kw={'xticks': [], 'yticks': []}, 
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, axi in enumerate(ax.flat):
    axi.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

```python
# 看看这些成分的累计方差，以及该投影保留了多少数据信息
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')    
```

- 可以看出这150个成分保留了95%以上的方差，这使我们相信，利用这150 个成分可以恢复数据的大部分必要特征

```python
# 计算成分和投影的人脸
pca = RandomizedPCA(150).fit(faces.data)      # 选择150个主成分训练数据
components = pca.transform(faces.data)        # 计算主成分
projected = pca.inverse_transform(components) # 将这些主成分通过逆变换重新构建投影数据（图像）

# 画出结果
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5), 
                      subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(faces.data[i].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i].reshape(62, 47), cmap='binary_r')

ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('150-dim\nreconstruction')
```

## 5.主成分分析PCA总结

- PCA 用途广泛、 可解释性强， 所以可以有效应用于大量情景和学科中
- 对于任意高维的数据集， 我倾向于以PCA 分析开始，包括：
    - 可视化点间的关系（正如手写数字示例中的处理方式）
    - 理解数据中的主要方差（正如特征脸示例中的处理方式）
    - 理解固有的维度（通过画出解释方差比） 
- 经常受数据集的异常点影响是 PCA 的主要弱点
    - 因为这个理由，很多效果更好的 PCA 变体被开发出来，这些 PCA 变体方法迭代执行，丢弃对原始成分描述得很糟糕的数据点
    - Scikit-Learn 中有一些PCA变体， 包括 RandomizedPCA 和 SparsePCA（这两个算法也在sklearn.decomposition 子模块中）
