```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
x = np.linspace(0, 10 ,1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), color='red')
```

```python
plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.sin(x - 1), color='c')
plt.plot(x, np.sin(x - 2)+1, color='0.75')
plt.plot(x, np.sin(x - 3), color='#ffdd55')
plt.plot(x, np.sin(x - 4)+2, color=(1, 0.3, 0.6))
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

```python
# 线条风格
plt.plot(x, x+1, linestyle='solid')
plt.plot(x, x+2, linestyle='dashed')
plt.plot(x, x+3, linestyle='dashdot')
plt.plot(x, x+4, linestyle='dotted')
plt.plot(x, x+5, linestyle='-')
plt.plot(x, x+6, linestyle='--')
plt.plot(x, x+7, linestyle='-.')
plt.plot(x, x+8, linestyle=':')
```

```python
# 调整坐标轴
plt.plot(x, np.cos(x))
plt.xlim(-0.1, 10)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x), '-.r')
# plt.axis([-0.1, 10, -1.2, 1.2])
# plt.axis('tight')
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x), '-.r')
plt.axis('tight')
plt.title('A Sin Curve', fontsize=20)
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '--c', label='sin(x)')
plt.plot(x, np.cos(x), '-.y', label='cos(x)')
plt.title('sin and cos curve', fontsize=15)
plt.axis('tight')
plt.xlabel('x')
plt.ylabel('sin(x)/cos(x)')
plt.legend(loc='lower left')
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x),':r')
ax.set(xlim=(-0.1,10), ylim=(-1.2,1.2),
      xlabel='x', ylabel='sin(x)',
      title='A SIN CURVE')
```
