```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-white')
```

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
print(x)
print('------------------------------------------------')
print(y)
print('------------------------------------------------')
print(X)
print('------------------------------------------------')
print(Y)
print('------------------------------------------------')
print(Z)
print('------------------------------------------------')
```

```python
plt.contour(X, Y, Z, colors='green')
```

```python
plt.contour(X, Y, Z, 20, cmap='RdYlBu')    # 可以用plt.cm.+<tab>来选择配色方案
```

```python
plt.figure(figsize=(7,5))
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
# plt.figure(figsize=(7,5))
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

```python
# 等高线图与彩色图叠放
plt.figure(figsize=(7,5))
contours = plt.contour(X, Y, Z, 5, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.contourf(X, Y, Z, 5, cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.savefig('../outputfiles/等高线图与彩色图叠放.jpg')
```

```python
# 等高线图与渐变的彩色图叠放
contours = plt.contour(X, Y, Z, 5, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
plt.savefig('../outputfiles/等高线图与渐变的彩色图叠放.jpg')
```
