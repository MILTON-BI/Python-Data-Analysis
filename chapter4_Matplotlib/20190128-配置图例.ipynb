```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '--b', label='Sin')
ax.plot(x, np.cos(x), '-r', label='Cos')
ax.axis('equal')
ax.le
```

```python
ax.legend(loc='upper left', frameon=False)
fig
```

```python
ax.legend(loc='lower center', frameon=True, ncol=2)
fig
```

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, ncol=2, loc='upper center', fontsize=12)
ax.set_title('sin and cos curve',fontsize=16)
fig
```

```python
# 选择图例显示的元素
# x = np.linspace(0, 10, 1000)
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)
print(lines)   
# [<matplotlib.lines.Line2D object at 0x000002DCAA1DE940>, 
# <matplotlib.lines.Line2D object at 0x000002DCAA1DEAC8>, 
# <matplotlib.lines.Line2D object at 0x000002DCAA1DEC18>, 
# <matplotlib.lines.Line2D object at 0x000002DCAA1DED68>]

# lines变量是一组plt.line2d实例
plt.legend(lines[:4], ['first', 'second', 'third', 'fourth'])
```
