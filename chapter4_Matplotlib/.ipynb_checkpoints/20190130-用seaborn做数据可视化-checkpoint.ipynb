```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

# 1.绘制简单的随机游走图

```python
# 创建随机游走数据
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

# 用matplotlib默认样式绘图
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

```python
# 用seaborn画上面的随机游走图像
import seaborn as sns
sns.set()
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

# 2. seaborn图形介绍

## 2.1 频次直方图、KDE和密度图

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density=True, alpha=0.5)
```

```python
# 通过KDE获取变量分布的平滑估计
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
# distplot将kde与频次直方图结合在一起
for col in 'xy':
     sns.distplot(data[col])
```

```python
# KDE输入二位数据集，可以得到二维的可视化图
sns.kdeplot(data['x'], data['y'])
```

```python
# sns.jointplot可以同时看到两个变量的联合分布，和单变量的独立分布
with sns.axes_style('white'):
    sns.jointplot('x', 'y', data, kind='kde')
```

```python
# 用六边形块代替频次直方图
with sns.axes_style('white'):
    sns.jointplot('x', 'y', data, kind='hex')
```

## 2.3 分面频次直方图

```python
tips = sns.load_dataset('tips', engine='python')
print(tips.head())
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
grid.map(plt.hist, 'tip_pct', bins=np.linspace(0, 40, 15))
```

## 2.4 因子图

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot('day', 'total_bill', 'sex', data=tips, kind="box")
    g.set_axis_labels('Day', 'Total Bill')
```

## 2.5 联合分布

```python
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip', data=tips, kind='hex')
```

```python
# 联合分布也可以自动进行KDE和回归
sns.jointplot('total_bill', 'tip', data=tips, kind='reg')
```

## 2.6 条形图

```python
# 时间序列数据可以用sns.factorplot画出条形图（用行星数据）
planets = sns.load_dataset('planets',engine='python')
print(planets.head())
```

```python
with sns.axes_style('white'):
    g = sns.factorplot('year', data=planets, aspect=2, kind='count', color='steelblue')
    g.set_xticklabels(step=2)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot('year', data=planets, aspect=4, kind='count', hue='method', order=range(2001, 2015))
    g.set_ylabels('Number of Planets Discovered')
```

# 案例：探索马拉松比赛成绩数据

```python
data = pd.read_csv('../data/marathon-data.csv')
print(data.head())
print(data.shape)    # (37250, 4)
print(data.dtypes)
```

```python
# 把字符串转变成时间类型的函数
import datetime
def convert_time(s):
    h, m, s = map(int, s.split(':'))  # 把时间字符串分割成时、分、秒
    return datetime.timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('../data/marathon-data.csv', converters={'split': convert_time, 'final': convert_time})
print(data.head())
print(data.dtypes)
```

```python
data['split_sec'] = data['split'] / np.timedelta64(1, 's')
data['final_sec'] = data['final'] / np.timedelta64(1, 's')
print(1E9)
print(data.head())
```

```python
# 用jointplot画图
with sns.axes_style('white'):
    g = sns.jointplot('split_sec', 'final_sec', data=data, kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000), np.linspace(8000, 32000), ':k')
```

```python
# 创建一列来表示选手的前后半程的差异系数（后半程加速或前半程加速）
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
print(data.head())
```

```python
# 差异系数分布图
sns.distplot(data['split_frac'], kde=False)
plt.axvline(0, color='k', linestyle='--')
```

```python
print(sum(data.split_frac < 0))   # 4万名选手中，只有251人在后半程加速（上图中的0点表示前后半程时间相同，>0表示前半程加速）
```

```python
# 前后半程差异系数与其他变量的相关性
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'], hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend()
```

```python
# 男女选手对比的频次直方图
sns.kdeplot(data.split_frac[data.gender=='M'], label='men', shade=True)
sns.kdeplot(data.split_frac[data.gender=='W'], label='men', shade=True)
plt.xlabel('split_frac')
```

```python
sns.violinplot('gender', 'split_frac', data=data, palette=['lightblue', 'lightpink'])
```

```python
# 创建一个新列表示年龄段
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
print(data.head())
```

```python
men = (data.gender=='M')
women = (data.gender=='W')
with sns.axes_style(style=None):
    sns.violinplot('age_dec', 'split_frac', hue='gender', data=data, split=True, inner='quartile', palette=['lightblue', 'lightpink'])
```

```python
# 后半程加速型选手的数据
g = sns.lmplot('final_sec', 'split_frac', col='gender', data=data, markers='.', scatter_kws=dict(color='c'))
g.map(plt.axhline, y=0.1, color='k', ls=":")
```
